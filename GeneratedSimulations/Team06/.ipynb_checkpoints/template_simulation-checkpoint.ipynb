## Defining all the the input parameters

```python
# Each value is a tuple with the following format: (mean, standard deviation)

import json

parameters = json.loads(open("parameters.json").read())
```

## Setting Up the Simulation

```python
env = Environment(
    railLength=parameters["railLength"],
    gravity=9.80665,
    date=datetime.strptime("2022-10-15 16:00:00", "%Y-%m-%d %H:%M:%S"),
    latitude=parameters["latitude"],
    longitude=parameters["longitude"],
    elevation=parameters["elevation"],
    datum="WSG84",
    timeZone="Portugal",
)
```

```python
env.setAtmosphericModel(type="Forecast", file="GFS")
```

```python
env.info()
```

```python
motor = SolidMotor(
    thrustSource="thrust_source.csv",
    burnOut=parameters["burnOut"],
    grainNumber=1,
    grainDensity=parameters["grainDensity"],
    grainOuterRadius=parameters["grainOuterRadius"],
    grainInitialInnerRadius=parameters["grainInitialInnerRadius"],
    grainInitialHeight=parameters["grainInitialHeight"],
    grainSeparation=0,
    nozzleRadius=parameters["nozzleRadius"],
    throatRadius=parameters["throatRadius"],
)
```

```python
motor.info()
```

```python
rocket = Rocket(
    motor=motor,
    radius=parameters["radius"],
    mass=parameters["rocketMass"],
    inertiaI=parameters["inertiaI"],
    inertiaZ=parameters["inertiaZ"],
    distanceRocketNozzle=parameters["distanceRocketNozzle"],
    distanceRocketPropellant=parameters["distanceRocketPropellant"],
    powerOffDrag="DragCoefficients.csv",
    powerOnDrag="DragCoefficients.csv",
)

# Add rocket nose, fins and tail

NoseCone = rocket.addNose(
    length=parameters["noseLength"],
    kind=parameters["noseShape"],
    distanceToCM=parameters["noseDistanceToCM"],
)

rocket.setRailButtons(
    distanceToCM=[parameters['railButtonDistToCM1'], 
                  parameters['railButtonDistToCM2']])
```

```python
for idx, fin in enumerate(filter(lambda x: "tailLength" in x, parameters.keys())):
    Tail = rocket.addTail(
      topRadius=parameters[f"tailTopRadius{idx}"],
      bottomRadius=parameters[f"tailBottomRadius{idx}"],
      length=parameters[f"tailLength{idx}"],
      distanceToCM=parameters[f"tailDistanceToCM{idx}"]
    )
```

```python
for idx, fin in enumerate(filter(lambda x: "finN" in x, parameters.keys())):
    FinSet = rocket.addTrapezoidalFins(
        n=parameters[f"finN{idx}"],
        span=parameters[f"finSpan{idx}"],
        rootChord=parameters[f"finRootChord{idx}"],
        tipChord=parameters[f"finTipChord{idx}"],
        distanceToCM=parameters[f"finDistanceToCm{idx}"],
        sweepLength=parameters.get(f"finSweepLength{idx}"),
        sweepAngle=parameters.get(f"finSweepAngle{idx}"),
        # radius=0,
        cantAngle=0,
        airfoil=None,
    )
```

```python
def drogueTrigger(p, y):
    # p = pressure
    # y = [x, y, z, vx, vy, vz, e0, e1, e2, e3, w1, w2, w3]
    # activate drogue when vz < 0 m/s.
    return True if y[5] < 0 else False


def mainTrigger(p, y):
    # p = pressure
    # y = [x, y, z, vx, vy, vz, e0, e1, e2, e3, w1, w2, w3]
    # activate main when vz < 0 m/s and z < 800 + main_parachute_trigger_altitude m
    # (+ main_parachute_trigger_altitude due to surface elevation).
    return (
        True
        if y[5] < 0 and y[2] < env.elevation + parameters[f"MainDeployAltitude0"]
        else False
    )
```

```python
if "DrogueDeployDelay1" in parameters.keys():
    print("This rocket might have more than one drogue")

if "MainDeployDelay1" in parameters.keys():
    print("This rocket might have more than one main")

# Add parachute
Drogue = rocket.addParachute(
    "Drogue",
    CdS=parameters[f"DrogueCds0"],
    trigger=drogueTrigger,
    samplingRate=105,
    lag=parameters[f"DrogueDeployDelay0"] + 0.001, # 0.001 adicionado para evitar warnings
    noise=(0, 8.3, 0.5),
)

Main = rocket.addParachute(
    "Main",
    CdS=parameters[f"MainCds0"],
    trigger=mainTrigger,
    samplingRate=105,
    lag=parameters[f"MainDeployDelay0"] + 0.001, # 0.001 adicionado para evitar warnings
    noise=(0, 8.3, 0.5),
)
```

```python
rocket.allInfo()
```

```python
TestFlight = Flight(
    rocket=rocket,
    environment=env,
    inclination=parameters["inclination"],
    heading=parameters["heading"],
)
```

## Saving the notebook into a .pdf file

```python
Flights = [(TestFlight, 'Nominal'), (TestFlightBallistic, 'Balistic'), (TestFlightOnlyDrogue, 'Drogue Only'), (TestFlightOnlyMain, 'Main Only')]

for TestFlight, Label in Flights:
    print(f"\n\n################################# FLIGHT MODE: {Label} ################################# \n")
    print(
    "Final Latitude: {:.5f} deg,\nFinal Longitude: {:.5f} deg".format(
        TestFlight.latitude(TestFlight.tFinal), TestFlight.longitude(TestFlight.tFinal)
        )
    )

    print(
        "Apogee Altitude: {:.3f} m (ASL) | {:.3f} m (AGL)".format(
            TestFlight.apogee, TestFlight.apogee - TestFlight.env.elevation
        )
    )
    print("Apogee Time: {:.3f} s".format(TestFlight.apogeeTime))
    print("Apogee Freestream Speed: {:.3f} m/s".format(TestFlight.apogeeFreestreamSpeed))

    # Print events registered
    print("\n\nEvents\n")
    if len(TestFlight.parachuteEvents) == 0:
        print("No Parachute Events Were Triggered.")
    for event in TestFlight.parachuteEvents:
        triggerTime = event[0]
        parachute = event[1]
        openTime = triggerTime + parachute.lag
        velocity = TestFlight.freestreamSpeed(openTime)
        altitude = TestFlight.z(openTime)
        name = parachute.name.title()
        print(name + " Ejection Triggered at: {:.3f} s".format(triggerTime))
        print(name + " Parachute Inflated at: {:.3f} s".format(openTime))
        print(
            name
            + " Parachute Inflated with Freestream Speed of: {:.3f} m/s".format(
                velocity
            )
        )
        print(
            name
            + " Parachute Inflated at Height of: {:.3f} m (AGL)".format(
                altitude - TestFlight.env.elevation
            )
        )
```
